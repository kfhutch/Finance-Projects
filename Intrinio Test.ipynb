### Working with Intrinio Data - Test Program 

```python
# Imports 

# IO Imports
import requests
import pprint
from pandas.io.json import json_normalize

# Data Imports
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# Code provided by Intrinio Customer Service to get started

# Credentials for Intrinio
with open('intriniocred.txt') as cred:
    api_username,api_password = cred.readlines()

api_username = api_username.strip()
api_password = api_password.strip()

# Stock ticker
ticker = "ATVI"

# Create list of dates
fundamentals_url = "https://www.intrinio.com/api/fundamentals/standardized"
p = {'ticker':ticker,'statement':'income_statement','type':'FY'}
r = requests.get(fundamentals_url, params=p, auth=(api_username, api_password))
date_list = r.json()['data']
print(len(date_list))

# Import data
statement_url = "https://www.intrinio.com/api/financials/standardized"

col_list = ['item']
test1 = False

for date_list_element in date_list:
    dd = date_list_element['end_date']
    col_list.append(dd)
    p = {'ticker':ticker,'statement':'income_statement','type':'FY','date':dd}
    r = requests.get(statement_url, params=p, auth=(api_username, api_password))
    if test1 == False:
        is_df = json_normalize(r.json()['data'])
        test1 = True
    else:
       is_df = is_df.merge(json_normalize(r.json()['data']),how='outer',on='tag',)

is_df.columns = [col_list]
is_df
```
